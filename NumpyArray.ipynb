```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.arange(1,101,1)
x  = x[:,np.newaxis]
Ones = np.ones(shape = (100,1))
x = np.hstack((x,np.square(x)))
x = np.hstack((x,Ones))
w = np.random.normal(loc = 0,scale = 1,size = (1,3))
weightage = np.array([[10,5,5]])
print(f"Initial Weightage:{w}!")
```

```python
y_pred = (w@(x.T)) .T
y_target = (weightage@(x.T)).T + np.random.normal(50,10,size = (100,1))
```

```python
y_target = y_target/y_target.max()
meanX = x.mean(axis = 0)
x = x/meanX
y_pred = y_pred/y_pred.mean()
```

```python
plt.scatter(x[:,0],y_target[:,0], label = "EXACT DATA")
plt.scatter(x[:,0],y_pred[:,0], label =  "PREDICTED DATA")
plt.legend()
```

```python
lamda = 0.0001
def loss_function(w,x,y_target):
    return (np.square((w@x.T).T - y_target)).sum()/(2*x.shape[0]) + lamda * (np.square(w)).sum()
```

```python
def grad_descent(w,x,y_target,alpha):
    dW  = + alpha*(((w@x.T).T - y_target)*x).sum(axis = 0 )/(x.shape[0]) + lamda*w
    # print(f"The change in weightage is {dW}")
    return dW
count = 0
while(count<100000):
    w = w -  grad_descent(w,x,y_target,0.01)
    y_pred = (w@x.T).T
    # print(f"Loss is {loss_function(w,x,y_target)} at weightage of {w}")
    count+=1
```

```python
plt.scatter(x[:,0],y_target[:,0], label = "EXACT DATA")
plt.scatter(x[:,0],y_pred[:,0], label =  "PREDICTED DATA")
plt.legend()
```
